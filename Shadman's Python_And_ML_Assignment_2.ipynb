1. Create a Car class with the following:
Attributes: brand, model, year, speed (default 0)
Methods: accelerate(amount) - increases speed, brake(amount) — decreases speed (min
0), and display_info() - prints all car details including current speed.
Create 2 car objects and call all methods on them.


```python
class Car:
  def __init__(self, brand, model, year, speed=0):
    self.brand = brand
    self.model = model
    self.year = year
    self.speed = speed

  def accelerate(self, amount):
    self.speed = self.speed + amount
    print("Speed increased by", amount)
    print("Current speed:", self.speed)

  def brake(self, amount):
    self.speed = self.speed - amount
    if self.speed < 0:
      self.speed = 0
    print("Speed decreased by", amount)
    print("Current speed:", self.speed)

  def display_info(self):
    print("Brand:", self.brand)
    print("Model:", self.model)
    print("Year:", self.year)
    print("Speed:", self.speed)

car1 = Car("Toyota", "Corolla", 2020)
car2 = Car("Honda", "Civic", 2022)

print("\nCar 1 Info:")
car1.display_info()
car1.accelerate(30)
car1.brake(10)
car1.display_info()
```

2. Create a BankAccount class with the following:
Attributes: account_holder, account_number, balance (default 0)
Methods: deposit(amount), withdraw(amount) - show error if balance is insufficient,
get_balance() - prints current balance, and transaction_history() - keeps a list of all
deposits and withdrawals and prints them.
Create 2 account objects and perform at least 3 transactions on each.

```python
class BankAccount:
  def __init__(self, account_holder, account_number, balance=0):
    self.account_holder = account_holder
    self.account_number = account_number
    self.balance = balance
    self.history = []

  def deposit(self, amount):
    self.balance = self.balance + amount
    self.history.append(f"Deposit: +{amount}")
    print("Deposit successful:", amount)
    print("Balance:", self.balance)

  def withdraw(self, amount):
    if amount > self.balance:
      print("Withdraw failed:", amount)
      print("Reason: Insufficient balance")
      print("Balance:", self.balance)
    else:
      self.balance = self.balance - amount
      self.history.append(f"Withdraw: -{amount}")
      print("Withdraw successful:", amount)
      print("Balance:", self.balance)

  def get_balance(self):
    print("Account Holder:", self.account_holder)
    print("Account Number:", self.account_number)
    print("Current Balance:", self.balance)

  def transaction_history(self):
    print("Transaction History for", self.account_holder)
    if len(self.history) == 0:
      print("No transactions yet.")
    else:
      for t in self.history:
        print(t)

acc1 = BankAccount("Sami", "001")
acc2 = BankAccount("Rahul", "002", 500)

print("\nAccount 1 Transactions:")
acc1.get_balance()
acc1.deposit(1000)
acc1.withdraw(200)
acc1.withdraw(900)
acc1.transaction_history()

print("\nAccount 2 Transactions:")
acc2.get_balance()
acc2.deposit(300)
acc2.withdraw(200)
acc2.withdraw(2000)
acc2.transaction_history()
```

3. Create a Library class and a Book class with the following:
Book attributes: title, author, isbn, is_available (default True)
Library attributes: name, books (empty list)
Library methods: add_book(book), borrow_book(isbn) - marks book unavailable,
return_book(isbn) - marks book available, and show_available_books() - prints all
available books.
Add at least 4 books to the library, borrow 2, return 1, and show available books.


```python
class Book:
  def __init__(self, title, author, isbn, is_available=True):
    self.title = title
    self.author = author
    self.isbn = isbn
    self.is_available = is_available

  def info(self):
    print("Title:", self.title, "| Author:", self.author, "| ISBN:", self.isbn, "| Available:", self.is_available)

class Library:
  def __init__(self, name):
    self.name = name
    self.books = []

  def add_book(self, book):
    self.books.append(book)
    print("Book added:", book.title)

  def borrow_book(self, isbn):
    found = False
    for b in self.books:
      if b.isbn == isbn:
        found = True
        if b.is_available == True:
          b.is_available = False
          print("Borrowed:", b.title)
        else:
          print("Not available:", b.title)
    if found == False:
      print("Book not found with ISBN:", isbn)

  def return_book(self, isbn):
    found = False
    for b in self.books:
      if b.isbn == isbn:
        found = True
        b.is_available = True
        print("Returned:", b.title)
    if found == False:
      print("Book not found with ISBN:", isbn)

  def show_available_books(self):
    print("Available Books in", self.name)
    any_book = False
    for b in self.books:
      if b.is_available == True:
        any_book = True
        b.info()
    if any_book == False:
      print("No available books right now.")

lib = Library("Bangla Sahitto Library")

lib.add_book(Book("Rabindranather Lipika", "Rabindranath", "978-81-247-0754-8"))
lib.add_book(Book("Gitanjali", "Rabindranath", "978-81-247-2454-5"))
lib.add_book(Book("Pather Panchali", "Bibhutibhushan", "978-81-247-1020-3"))
lib.add_book(Book("Aparajito", "Bibhutibhushan", "978-81-247-1165-1"))
lib.add_book(Book("Devdas", "Sarat Chandra", "978-81-247-1402-7"))
lib.add_book(Book("Shesher Kobita", "Rabindranath", "978-81-247-1710-3"))

print("\nBorrow 2 books:")
lib.borrow_book("978-81-247-0754-8")
lib.borrow_book("978-81-247-1402-7")

print("\nReturn 1 book:")
lib.return_book("978-81-247-0754-8")

print("\nShow available books:")
lib.show_available_books()
```

4. Create an Employee class with the following:
Attributes: name, employee_id, department, salary
Methods: give_raise(percent) - increases salary by the given percentage, details() - prints
all employee info, and a class variable total_employees that tracks how many Employee
objects have been created.
Create 3 employees, give raises to 2 of them, and print total_employees using the class
name.


```python
class Employee:
  total_employees = 0

  def __init__(self, name, employee_id, department, salary):
    self.name = name
    self.employee_id = employee_id
    self.department = department
    self.salary = salary
    Employee.total_employees = Employee.total_employees + 1

  def give_raise(self, percent):
    increase = self.salary * (percent / 100)
    self.salary = self.salary + increase
    print(self.name, "got a raise of", percent, "%")
    print("New salary:", self.salary)

  def details(self):
    print("Name:", self.name)
    print("ID:", self.employee_id)
    print("Department:", self.department)
    print("Salary:", self.salary)

emp1 = Employee("Rafi", "E-01", "Sales", 30000)
emp2 = Employee("Nila", "E-02", "IT", 50000)
emp3 = Employee("Joy", "E-03", "HR", 35000)

print("\nEmployee Details:")
emp1.details()
print()
emp2.details()
print()
emp3.details()

print("\nGive raise to 2 employees:")
emp1.give_raise(10)
emp2.give_raise(20)

print("\nTotal employees:", Employee.total_employees)
```

5. Create a ShoppingCart class with the following:
Attributes: customer_name, items (empty list - each item is a dict with name, price,
quantity)
Methods: add_item(name, price, quantity), remove_item(name), total_price() - returns
total cost of all items, and show_cart() - prints all items with their details and final total.
Create 2 shopping cart objects for different customers, add at least 3 items each, remove
1 item from each, and show the cart with total.

```python
class ShoppingCart:
  def __init__(self, customer_name):
    self.customer_name = customer_name
    self.items = []

  def add_item(self, name, price, quantity):
    self.items.append({"name": name, "price": price, "quantity": quantity})
    print("Item added for", self.customer_name, ":", name)

  def remove_item(self, name):
    removed = False
    new_list = []
    for it in self.items:
      if it["name"] == name and removed == False:
        removed = True
      else:
        new_list.append(it)
    self.items = new_list
    if removed == True:
      print("Item removed for", self.customer_name, ":", name)
    else:
      print("Item not found for", self.customer_name, ":", name)

  def total_price(self):
    total = 0
    for it in self.items:
      total = total + (it["price"] * it["quantity"])
    return total

  def show_cart(self):
    print("Cart for:", self.customer_name)
    if len(self.items) == 0:
      print("No items in cart.")
    else:
      for it in self.items:
        print("Name:", it["name"], "| Price:", it["price"], "| Qty:", it["quantity"], "| Subtotal:", it["price"] * it["quantity"])
      print("Final Total:", self.total_price())

cart1 = ShoppingCart("Sameer")
cart2 = ShoppingCart("Ayesha")

print("\nCart 1 Add Items:")
cart1.add_item("Rice", 80, 2)
cart1.add_item("Milk", 60, 3)
cart1.add_item("Eggs", 12, 10)

print("\nCart 2 Add Items:")
cart2.add_item("Bread", 50, 2)
cart2.add_item("Juice", 120, 1)
cart2.add_item("Apples", 30, 8)

print("\nRemove 1 item from each cart:")
cart1.remove_item("Milk")
cart2.remove_item("Juice")

print("\nShow carts:")
cart1.show_cart()
print()
cart2.show_cart()
```
