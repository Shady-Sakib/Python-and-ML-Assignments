(a) Age Checker — take the user's age as input:

• Age > 18 → print 'You are an adult'
• Age 13–17 → print 'You are a teenager'
• Age < 13 → print 'You are a child'
• Also handle: if the age entered is negative → print 'Invalid age'

```python
user_age = int(input("Enter your age: "))

if user_age < 0:
	print("Invalid age")
elif user_age > 18:
	print("You are an adult")
elif 13 <= user_age <= 17:
	print("You are a teenager")
else:
	print("You are a child")
```

(c) Grade Calculator — take a student's mark (0–100) as input:

• 50–59 → 'D'

(d) Login System — simulate a simple login:

• Store a correct username and password in your code (e.g., username = "admin", password =
"1234")
• Ask the user to enter their username and password
• If both match → print 'Login successful! Welcome, [username].'
• If username is correct but password is wrong → print 'Incorrect password. Try again.'
• If username is wrong → print 'User not found.'
• Give the user a maximum of 3 attempts. After 3 failed tries → print 'Account locked.'
• Hint: Use a while loop with a counter for the attempt logic.

```python
correct_username = "admin"
correct_password = "1234"

attempts = 3
logged_in = False

while attempts > 0 and logged_in == False:
	u = input("Enter username: ")
	p = input("Enter password: ")

	if u == correct_username and p == correct_password:
		print(f"Login successful! Welcome, {u}.")
		logged_in = True
	elif u == correct_username and p != correct_password:
		attempts -= 1
		if attempts == 0:
			print("Account locked.")
		else:
			print("Incorrect password. Try again.")
	else:
		attempts -= 1
		if attempts == 0:
			print("Account locked.")
		else:
			print("User not found.")
```

Attempts remaining: 2')

(c) Add a Login System before the ATM:

• Ask for a username and password before the ATM section
• Add a variable is_active = True — if False, print 'Account is disabled.' and stop
• Only proceed to the ATM pin check if login is successful

```python
account_username = "admin"
account_password = "1234"
is_active = True

if is_active == False:
	print("Account is disabled.")
else:
	u2 = input("Enter username for ATM: ")
	p2 = input("Enter password for ATM: ")

	if u2 != account_username or p2 != account_password:
		print("Login failed. Cannot proceed to ATM.")
	else:
		pinNumber = "2468"
		balance = 5000
		attempts = 3
		pin_ok = False

		while attempts > 0 and pin_ok == False:
			entered_pin = input("Enter ATM PIN: ")
			if entered_pin == pinNumber:
				pin_ok = True
			else:
				attempts -= 1
				if attempts == 0:
					print("Card blocked!")
				else:
					print(f"Wrong pin. Attempts remaining: {attempts}")

		if pin_ok == True:
			request = int(input("Enter withdrawal amount: "))

			if request <= 500:
				print("Withdrawal must be greater than 500.")
			elif request > balance:
				print("Insufficient balance.")
			else:
				balance -= request
				print("Withdrawal successful.")
				print("Remaining balance:", balance)
```

(a) for loop — basic practice:

• Print numbers from 1 to 20 using a for loop
• Print the multiplication table of 7 (7×1 up to 7×10)
• Calculate and print the sum of all numbers from 1 to 100
• Print all even numbers between 1 and 50 using range() with a step
• Given the list fruits = ['apple', 'banana', 'cherry', 'mango', 'grape'], loop through it and print each
fruit with its position number (starting from 1)

```python
print("Numbers from 1 to 20:")
for i in range(1, 21):
	print(i)

print("\nMultiplication table of 7:")
for i in range(1, 11):
	print(f"7 x {i} = {7 * i}")

total_sum = 0
for i in range(1, 101):
	total_sum += i
print("\nSum of 1 to 100:", total_sum)

print("\nEven numbers between 1 and 50:")
for i in range(2, 51, 2):
	print(i)

fruits = ['apple', 'banana', 'cherry', 'mango', 'grape']
print("\nFruits with position:")
pos = 1
for f in fruits:
	print(pos, f)
	pos += 1
```

(b) while loop — basic practice:

• Print numbers from 10 down to 1 using a while loop
• Keep asking the user to enter a positive number. Stop (using break) when they enter 0 or a negative
number. Print the total sum of all valid numbers entered.
• Simulate a countdown: start from 10, print each number, then print 'Blast off!' when it reaches 0
• Find and print the first 10 multiples of 6 using a while loop

```python
print("Numbers from 10 down to 1:")
n = 10
while n >= 1:
	print(n)
	n -= 1

print("\nEnter positive numbers (stop on 0 or negative).")
sum_valid = 0
while True:
	val = int(input("Enter a number: "))
	if val <= 0:
		break
	sum_valid += val
print("Total sum of valid numbers:", sum_valid)

print("\nCountdown to Blast off!")
c = 10
while c >= 0:
	print(c)
	if c == 0:
		print("Blast off!")
	c -= 1

print("\nFirst 10 multiples of 6:")
count = 1
while count <= 10:
	print(6 * count)
	count += 1
```
